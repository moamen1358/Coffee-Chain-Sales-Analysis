# what is the product that has the highest profit

```python
data1 = df[['Product','Profit']].groupby('Product').sum().reset_index().sort_values(by='Profit')
data1
```

```python
ax = sns.barplot(data= data1, x='Product', y='Profit', color='r')
sns.despine(left=True, bottom=True)
plt.xticks(rotation=90)
plt.yticks([],[])
ax.bar_label(ax.containers[0]);
```

# what is the highest sales by year

```python
df.head()
```

```python
df['Year'] = df['Date'].apply(lambda x : x.split('/')[2])
```

```python
df['Year'] = pd.to_numeric(df['Year'])
```

```python
data2 = df[['Year', 'Sales']].groupby('Year').sum().reset_index()
data2
```

```python
ax = sns.barplot(data= data2, x='Year' , y='Sales', color='green')
sns.despine(left=True, bottom=True)
plt.yticks([],[])
ax.bar_label(ax.containers[0]);
```
